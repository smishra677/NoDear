```python
import allel
import numpy as np
taj_eq=[]
for i in range(120):
    ha = np.load('./eq_tes/'+str(i)+'_haps.npy')


    h = allel.HaplotypeArray(ha)


    g = h.to_genotypes(ploidy=1) 

    ac = g.count_alleles()

    tajimas_d = allel.tajima_d(ac)
    taj_eq.append(tajimas_d)

```

```python
def make_shannon(H):
    from scipy.stats import entropy
    li=[]
    lis=[]

    for i in H:
        co = dict(Counter(i))
        if len(co.keys())==2:
                lis.append(co[min(co,key=co.get)]/len(i))
    bins = np.linspace(0, 1, 20)

    bin_means,bins = np.histogram(lis, bins, normed=True, density=True)
    bins= bins[1:]
    bin_means= bin_means/bin_means.sum()
    
    print(bin_means)

   
    li.append(entropy(bin_means[:10]))
    
    return li[0]
        
```

```python

import numpy as np
ent_conv=[]
ent_non_eq=[]
ent_eq=[]
ent_split=[]
for i in range(120):
    ha_conv = np.load('gene_conversion_eq/'+str(i)+'_haps.npy')
    ha_non_eq = np.load('non_eq_tes/'+str(i)+'_haps.npy')
    ha_eq = np.load('eq_tes/'+str(i)+'_haps.npy')
    ha_split = np.load('pop_split/'+str(i)+'_haps.npy')

    ent = make_shannon(ha_conv)
    ent_conv.append(ent)

    ent = make_shannon(ha_non_eq)
    ent_non_eq.append(ent)


    ent = make_shannon(ha_eq)
    ent_eq.append(ent)


    ent = make_shannon(ha_split)
    ent_split.append(ent)

```

```python
print('split', 'mean',np.array(ent_split).mean(),'var', np.array(ent_split).var())
print('non_eq', 'mean',np.array(ent_non_eq).mean(),'var',  np.array(ent_non_eq).var())
print('eq', 'mean',np.array(ent_eq).mean(),'var', np.array(ent_eq).var())
print('50%conv', 'mean',np.array(ent_conv).mean(),'var', np.array(ent_conv).var())
```
